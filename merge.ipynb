```python
import pandas as pd
import numpy as np
```

```python
from datetime import datetime
```

```python
df = pd.read_csv('2021-02-23-12-18-09-EST-Historical-Report-SP-500-2020-01-01--2021-02-24.csv')
```

```python
df
```

```python
from dateutil import parser
```

```python
df.Created = df.Created.apply(lambda x: parser.parse(x))
```

```python
result = df.groupby(['User Name'])['Facebook Id'].count()
```

```python
result = result.to_frame()
```

```python
result_2 = df.groupby(['User Name'])['Likes','Comments','Shares','Love','Wow','Haha','Sad','Angry','Care'].sum()
```

```python
result = result.merge(result_2, right_index =True, left_index = True)
```

```python
result = result.reset_index()
```

```python
def get_date(x):
    global df
    temp = df[df['User Name'] == x]
    temp_sort = temp.sort_values(by = ['Created'])
    sd = temp_sort.Created.iloc[0]
    ed = temp_sort.Created.iloc[-1]
    return sd, ed
```

```python
result['start_date'] = result['User Name'].apply(lambda x:get_date(x)[0] )
```

```python
result['end_date'] = result['User Name'].apply(lambda x:get_date(x)[1] )
```
